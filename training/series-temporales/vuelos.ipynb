# Análisis de Series Temporales con Datos Reales: AirPassengers

Este cuaderno analiza datos reales de pasajeros aéreos mensuales en EE.UU. entre 1949 y 1960. Se explican y calculan el **MAE** (Mean Absolute Error) y el **RMSE** (Root Mean Squared Error), además de ilustrar su uso en series temporales.

## 1. ¿Qué son el MAE y el RMSE?

**Mean Absolute Error (MAE)** mide el error absoluto medio entre los valores reales y predichos:
$$ MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i| $$

Donde:
- \( y_i \) es el valor real.
- \( \hat{y}_i \) es el valor predicho.
- \( n \) es el número de observaciones.

El **MAE** mide el error en las mismas unidades de los datos originales y es fácil de interpretar.

**Root Mean Squared Error (RMSE)** mide la raíz cuadrada del error cuadrático medio:
$$ RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2} $$

El **RMSE** penaliza más los errores grandes debido a la elevación al cuadrado, por lo que es útil cuando queremos minimizar grandes desviaciones en nuestras predicciones.

### Ejemplo Ilustrativo:
Imagina que queremos predecir la temperatura diaria de una ciudad y nuestro modelo predice estos valores:
| Día | Temperatura Real (°C) | Predicción (°C) |
|-----|----------------------|----------------|
| 1   | 20                   | 21             |
| 2   | 22                   | 24             |
| 3   | 19                   | 18             |

El **MAE** nos dirá, en promedio, cuánto nos equivocamos en cada predicción, mientras que el **RMSE** nos indicará cuánto afectan los errores grandes.

## 2. Importar librerías necesarias

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

## 3. Cargar y visualizar el dataset AirPassengers

Este conjunto de datos contiene el número de pasajeros aéreos mensuales en los EE.UU. entre 1949 y 1960.

```python
df = pd.read_csv('vuelos.csv')
df
```

## Explorando Datos Temporales: Patrones en Vuelos

Antes de métricas, **explora los datos**. Los pasajeros aéreos muestran tendencias (aumento general) y estacionalidad (más vuelos en verano).

### ¿Por qué explorar?
- **Revela Tendencias**: ¿Más pasajeros con el tiempo? ¿Ciclos mensuales?

- **Prepara Evaluación**: Entiende el contexto antes de medir errores.

## 4. Visualizar la serie temporal de pasajeros aéreos

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=df['Date'], y=df['Passengers'], marker='o')
plt.xlabel("Fecha")
plt.ylabel("Número de Pasajeros")
plt.title("Número de Pasajeros Aéreos por Mes (1949-1960)")
plt.show()
```

## 5. Crear una predicción con promedio móvil

Para fines educativos, usaremos un método simple: el promedio móvil de 12 meses.

```python
df['Passengers_Predicted'] = df['Passengers'].shift(1).rolling(window=12).mean()
df
```

## Métricas de Evaluación: MAE y RMSE en Acción

Ahora, calculamos **MAE** y **RMSE** para medir errores en predicciones. Imagina predecir pasajeros: MAE dice el error promedio; RMSE, cuánto duelen los grandes errores.

### Conceptos clave:
- **MAE**: Promedio de diferencias absolutas. Fácil, en unidades originales.
- **RMSE**: Raíz de promedio cuadrático. Penaliza outliers, sensible a grandes errores.
- **Cuándo usar**: MAE para interpretabilidad; RMSE para minimizar desviaciones grandes.

### Lección educativa:
- Calcula siempre ambas: Compara modelos.
- Bajo es mejor, pero contextualiza: ¿Qué error es aceptable en aviación?
- Experimenta: Simula predicciones y ve cómo cambian métricas.

¡Mide el error y mejora tus modelos!

## 6. Calcular MAE y RMSE

Ahora calculamos estos errores para evaluar la calidad de la predicción.

```python
# Eliminar los primeros 12 valores nulos debido al promedio móvil
mae = mean_absolute_error(df['Passengers'][12:], df['Passengers_Predicted'][12:])
rmse = np.sqrt(mean_squared_error(df['Passengers'][12:], df['Passengers_Predicted'][12:]))

print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
```

## 7. Visualización de la predicción

Graficamos los valores reales y los predichos para ver la diferencia.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=df['Date'], y=df['Passengers'], label='Pasajeros Reales', marker='o')
sns.lineplot(x=df['Date'], y=df['Passengers_Predicted'], label='Predicción (Prom. Móvil)', linestyle='dashed')
plt.xlabel("Fecha")
plt.ylabel("Número de Pasajeros")
plt.title("Predicción de Pasajeros Aéreos con Promedio Móvil")
plt.legend()
plt.show()
```
