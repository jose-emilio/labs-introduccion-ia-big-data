# Análisis de Series Temporales: Temperaturas en Melbourne

Este cuaderno analiza datos reales de temperaturas mínimas diarias en Melbourne, Australia, desde 1981 hasta 1990. Se aplicará un modelo de **promedio móvil** para realizar predicciones y se evaluará su desempeño utilizando **MAE (Mean Absolute Error)** y **RMSE (Root Mean Squared Error)**.

## 1. Importar librerías necesarias

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

## 2. Cargar y visualizar el dataset

El conjunto de datos contiene las temperaturas mínimas diarias en Melbourne desde 1981 hasta 1990. Lo descargamos y convertimos la columna de fecha a formato datetime.

```python
# Cargar dataset real de temperaturas mínimas diarias en Melbourne
df = pd.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv")
df.columns = ['Date', 'Temperature']
df['Date'] = pd.to_datetime(df['Date'])  # Convertir a formato de fecha
df.to_csv('temperaturas.csv', index=False)
```

```python
df = pd.read_csv('temperaturas.csv')
df
```

## ¿Qué es el Análisis Exploratorio en Series Temporales?

Antes de modelar, **explora tus datos temporales**. Las series temporales tienen patrones únicos: tendencias (aumento/disminución), estacionalidad (repetitivo, como estaciones) y ruido (aleatorio).

### ¿Por qué es crucial?
- **Detecta Patrones**: ¿Las temperaturas suben en verano? ¿Hay ciclos anuales?
- **Identifica Problemas**: Datos faltantes o outliers pueden arruinar predicciones.
- **Guía el Modelo**: Elige el modelo correcto basado en lo que ves.

### Tips educativos:
- Siempre grafica la serie completa: Usa líneas para ver evolución.
- Descompón: Separa tendencia, estacionalidad y residuales.
- Piensa en el contexto: Temperaturas en Melbourne varían por hemisferio sur.

Recuerda: Una buena exploración revela historias ocultas en el tiempo.

## 3. Visualización inicial de la serie temporal

A continuación, graficamos la serie completa para observar su comportamiento.

```python
plt.figure(figsize=(14, 6))
sns.lineplot(x=df['Date'], y=df['Temperature'])
plt.xlabel("Fecha")
plt.ylabel("Temperatura Mínima (°C)")
plt.title("Temperaturas Mínimas Diarias en Melbourne (1981-1990)")
plt.show()
```

## 4. Análisis de estacionalidad y tendencias

Podemos visualizar mejor los patrones estacionales promediando por mes y por año.

```python
# Extraer año y mes para análisis estacional
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

# Promedio mensual para ver estacionalidad
monthly_avg = df.groupby('Month')['Temperature'].mean()

plt.figure(figsize=(10, 5))
sns.barplot(x=monthly_avg.index, y=monthly_avg.values)
plt.xlabel("Mes")
plt.ylabel("Temperatura Promedio (°C)")
plt.title("Promedio de Temperatura Mínima por Mes (1981-1990)")
plt.show()
```

## Introducción a Modelos de Series Temporales

¡Ahora modelamos! Usaremos un **promedio móvil**, que predice basándose en el promedio de valores recientes. Es simple pero efectivo para tendencias suaves.

### Conceptos clave:
- **Promedio Móvil**: Calcula el promedio de los últimos N puntos para predecir el siguiente.
- **Ventana (Window)**: El número de puntos (e.g., 7 días para una semana).
- **Evaluación**: MAE (error absoluto medio) y RMSE (raíz del error cuadrático medio) miden precisión.
- **Ventajas/Desventajas**: Rápido, pero ignora tendencias complejas.

### Lección educativa:
- Modelos simples son un buen inicio; compara con complejos como ARIMA.
- Evalúa siempre: Un bajo error no significa buen modelo si no generaliza.
- Experimenta: Cambia la ventana y ve cómo afecta predicciones.

¡Ve cómo un cálculo básico predice temperaturas!

## 5. Aplicar un modelo de predicción con Promedio Móvil

Usamos un promedio móvil de 30 días para suavizar la serie y hacer predicciones.

```python
df['Temperature_Predicted'] = df['Temperature'].shift(1).rolling(window=30).mean()
df.head(35)
```

## 6. Evaluar la precisión con MAE y RMSE

Calculamos estos errores para medir la calidad de la predicción.

```python
# Eliminar los primeros 30 valores nulos debido al promedio móvil
mae = mean_absolute_error(df['Temperature'][30:], df['Temperature_Predicted'][30:])
rmse = np.sqrt(mean_squared_error(df['Temperature'][30:], df['Temperature_Predicted'][30:]))

print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
```

## 7. Visualización de la predicción

Graficamos los valores reales y los predichos con el promedio móvil.

```python
plt.figure(figsize=(14, 6))
sns.lineplot(x=df['Date'], y=df['Temperature'], label='Temperatura Real', alpha=0.6)
sns.lineplot(x=df['Date'], y=df['Temperature_Predicted'], label='Predicción (Prom. Móvil)', linestyle='dashed')
plt.xlabel("Fecha")
plt.ylabel("Temperatura Mínima (°C)")
plt.title("Predicción de Temperaturas Mínimas con Promedio Móvil (30 días)")
plt.legend()
plt.show()
```

## 8. Conclusión

En este cuaderno, analizamos una serie temporal de temperaturas mínimas en Melbourne y aplicamos un modelo de predicción basado en promedio móvil. Se observó que este método suaviza la serie pero con un retraso evidente en la predicción. El MAE y RMSE nos ayudan a cuantificar el error.
