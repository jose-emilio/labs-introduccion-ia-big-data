# Paso 3: Limpieza de Datos - Preparando Mediciones Ambientales Confiables

## Desafíos de la Limpieza en Datos Ambientales

Los datos de calidad del aire, como muchos datasets científicos, presentan desafíos únicos en su limpieza. Los valores faltantes no son meros errores, sino reflejos de limitaciones instrumentales y condiciones ambientales reales que deben manejarse con cuidado estadístico y científico.

## Naturaleza de los Valores Faltantes en Datos Atmosféricos

### Causas Comunes de Datos Faltantes

1. **Condiciones Meteorológicas Adversas**
   - Lluvia intensa que daña sensores.
   - Tormentas que interrumpen mediciones.
   - Niebla densa que afecta lecturas ópticas.

2. **Problemas Técnicos**
   - Fallos en calibración de instrumentos.
   - Mantenimiento programado de equipos.
   - Interrupciones en suministro eléctrico.

3. **Limitaciones de Detección**
   - Concentraciones por debajo del límite de detección.
   - Sobrepaso de rango instrumental.
   - Interferencias en mediciones.

4. **Factores Humanos**
   - Errores en registro manual.
   - Problemas en transmisión de datos.
   - Cambios en protocolos de medición.

## Estrategia de Imputación: Uso de la Mediana

### ¿Por Qué la Mediana?

En datos ambientales, la mediana ofrece ventajas significativas sobre la media:

- **Robustez a outliers**: No se ve afectada por valores extremos.
- **Representatividad**: Mejor estimador para distribuciones sesgadas.
- **Conservadurismo**: Mantiene la estructura estadística original.

### Variables que Reciben Imputación

- **Ozone**: Concentraciones que pueden tener outliers por eventos de smog.
- **Solar.R**: Radiación que varía con nubosidad.
- **Wind**: Velocidad del viento, generalmente simétrica.
- **Temp**: Temperatura, con distribución relativamente normal.

### Implementación Técnica

Utilizaremos `df.fillna(df.median(numeric_only=True), inplace=True)`:

- **numeric_only=True**: Asegura solo variables numéricas.
- **inplace=True**: Modifica el DataFrame original.
- **Método**: Imputación por columna con mediana respectiva.

## Consideraciones Estadísticas

### Impacto en la Distribución

- **Mediana preservada**: La medida central no cambia.
- **Dispersión**: Ligero aumento en varianza.
- **Correlaciones**: Mínimo impacto en relaciones entre variables.

### Validación de Imputación

Después de la imputación, verificaremos:

- **Reducción de nulos**: Confirmar eliminación de valores faltantes.
- **Distribuciones**: Comparar antes y después.
- **Lógica ambiental**: Valores imputados plausibles.

## Alternativas Consideradas

### Otras Estrategias de Imputación

1. **Media aritmética**: Simple pero sensible a outliers.
2. **Interpolación temporal**: Apropiada para series de tiempo.
3. **Modelos predictivos**: KNN o regresión para imputación.
4. **Eliminación**: Solo si faltan pocos datos.

### Justificación de Elección

La mediana fue elegida porque:

- **Robustez**: Datos ambientales tienen outliers naturales.
- **Simplicidad**: Método transparente y reproducible.
- **Efectividad**: Funciona bien con distribuciones sesgadas.
- **Aceptación**: Estándar en literatura ambiental.

## Implicaciones para el Análisis Ambiental

### Efectos en Interpretaciones

- **Tendencias**: Imputación puede suavizar variabilidad real.
- **Umbrales**: Valores imputados pueden afectar conteos de días "malos".
- **Modelos predictivos**: Imputación introduce incertidumbre.

### Mejores Prácticas

- **Documentación**: Registrar método y justificación.
- **Sensibilidad**: Probar análisis con diferentes estrategias.
- **Transparencia**: Indicar variables imputadas en reportes.

## Implementación en Código

En la siguiente celda, aplicaremos:

# Paso 4: Transformación de Datos - Normalizando y Categorizando Variables Ambientales

## Ingeniería de Características en Datos Atmosféricos

La transformación de datos en estudios ambientales va más allá de la limpieza; implica crear representaciones que capturen mejor las complejidades atmosféricas y faciliten análisis predictivos y comparativos.

## Transformaciones Específicas Aplicadas

### 1. Normalización Min-Max del Ozono

**Variable Nueva**: `Ozone_Normalizado`

**Justificación Científica**:
- **Escalado consistente**: Permite comparación directa entre días.
- **Interpretación relativa**: Facilita identificación de días con ozono alto/bajo.
- **Preparación para modelos**: Algoritmos sensibles a escala (redes neuronales, KNN).

**Metodología Técnica**:
- **Fórmula**: `(valor - mínimo) / (máximo - mínimo)`
- **Rango resultante**: [0, 1]
- **Preservación de forma**: Mantiene distribución relativa.

**Aplicaciones Ambientales**:
- **Monitoreo comparativo**: Días con ozono > 0.8 son críticos.
- **Análisis de tendencias**: Cambios relativos en concentraciones.
- **Modelado predictivo**: Variable escalada para algoritmos ML.

### 2. Categorización de Temperatura

**Variable Nueva**: `Categoria_Temperatura`

**Justificación Científica**:
- **Clasificación meteorológica**: Temperaturas en categorías significativas.
- **Análisis estacional**: Facilita comparación entre períodos.
- **Interpretación intuitiva**: Categorías más comprensibles que números.

**Metodología Técnica**:
- **Método**: `pd.cut()` con bins discretos.
- **Categorías**: 'Fresca' (50-70°F), 'Cálida' (70-90°F), 'Calurosa' (90-110°F)
- **Inclusión**: right=False para límites superiores exclusivos.

**Aplicaciones Ambientales**:
- **Estudios de salud**: Temperaturas calurosas correlacionan con hospitalizaciones.
- **Análisis de contaminación**: Ozono aumenta con temperaturas altas.
- **Planificación urbana**: Días calurosos requieren medidas especiales.

## Consideraciones Técnicas Avanzadas

### Normalización Min-Max: Ventajas y Limitaciones

**Ventajas**:
- **Intuitiva**: Valores entre 0 y 1 fáciles de interpretar.
- **Preserva relaciones**: Proporcionalidad mantenida.
- **Independiente de unidades**: Comparable entre variables.

**Limitaciones**:
- **Sensible a outliers**: Valores extremos distorsionan escala.
- **No robusta**: Un solo outlier cambia toda la normalización.
- **Asume distribución**: Mejor para datos uniformemente distribuidos.

### Categorización: Estrategias y Justificaciones

**Elección de Bins**:
- **50-70°F**: Rango fresco, típico de primavera/temprano otoño.
- **70-90°F**: Rango cálido, veranos moderados.
- **90-110°F**: Rango caluroso, olas de calor.

**Criterios**:
- **Relevancia ambiental**: Basado en estándares meteorológicos.
- **Equilibrio**: Categorías aproximadamente igualadas.
- **Interpretabilidad**: Nombres descriptivos en español.

## Impacto en el Análisis Ambiental

### Mejoras en Interpretabilidad

- **Ozono normalizado**: Facilita comparación temporal y espacial.
- **Categorías de temperatura**: Simplifica análisis estacionales.

### Preparación para Modelado Avanzado

- **Variables escaladas**: Listas para algoritmos de machine learning.
- **Características categóricas**: Preparadas para encoding (one-hot, ordinal).

### Implicaciones Científicas

- **Correlaciones**: Temperatura categórica puede revelar patrones no lineales.
- **Umbrales**: Valores normalizados ayudan en definición de alertas.
- **Comparabilidad**: Datos estandarizados permiten meta-análisis.

## Validación de Transformaciones

Después de aplicar, verificaremos:

- **Rangos correctos**: Ozono normalizado ∈ [0,1].
- **Categorías apropiadas**: Temperaturas asignadas correctamente.
- **Distribuciones preservadas**: Forma relativa mantenida.
- **Valores faltantes**: Ninguno introducido por transformación.

## Extensiones Futuras

Estas transformaciones básicas pueden expandirse:

- **Normalización Z-score**: Para datos con outliers.
- **Categorías dinámicas**: Basadas en percentiles.
- **Variables derivadas**: Índices de calidad del aire compuestos.
- **Transformaciones no lineales**: Logaritmos para variables sesgadas.

## Implementación en Código

En la siguiente celda, ejecutaremos:

1. **Normalización de ozono**: Aplicación de min-max scaling.
2. **Categorización de temperatura**: Creación de bins discretos.
3. **Verificación**: Inspección del dataset transformado.

## Integración con Análisis Posteriores

Estas nuevas variables enriquecerán:

- **Visualizaciones**: Gráficos más informativos.
- **Modelos predictivos**: Variables mejor preparadas.
- **Interpretaciones**: Insights más claros sobre dinámica atmosférica.

Esta transformación marca la evolución de datos limpios a datos enriquecidos, preparados para revelar los complejos patrones de la calidad del aire en Nueva York de 1973.

```python
# Paso 4: Transformación de datos
# Crear una nueva columna con la normalización min-max de la concentración de ozono (Ozone)
if 'Ozone' in df.columns:
    df['Ozone_Normalizado'] = (df['Ozone'] - df['Ozone'].min()) / (df['Ozone'].max() - df['Ozone'].min())

# Crear una nueva columna categórica basada en la temperatura (Temp)
if 'Temp' in df.columns:
    df['Categoria_Temperatura'] = pd.cut(df['Temp'], bins=[50, 70, 90, 110], labels=['Fresca', 'Cálida', 'Calurosa'], right=False)

print("\nDatos después de la transformación:")
df
```
