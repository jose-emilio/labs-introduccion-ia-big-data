- **Codificación de variables categóricas**: Conversión de variables no numéricas a formatos numéricos.
- **Ingeniería de características**: Creación de nuevas variables derivadas de las existentes.

# Paso 3: Limpieza de Datos - Preparando el Dataset para Análisis Avanzados

## Fundamentos de la Limpieza de Datos

La limpieza de datos es uno de los aspectos más críticos del preprocesamiento. Los datos del mundo real rara vez están en un formato perfecto; suelen contener errores, inconsistencias y valores faltantes que pueden comprometer la calidad de cualquier análisis o modelo predictivo.

### ¿Por Qué es Importante la Limpieza de Datos?

- **Mejora la precisión**: Datos limpios producen resultados más confiables.
- **Reduce el sesgo**: Elimina inconsistencias que podrían sesgar los análisis.
- **Optimiza el rendimiento**: Algoritmos funcionan mejor con datos consistentes.
- **Facilita la interpretación**: Resultados más claros y comprensibles.

## Estrategias de Tratamiento de Valores Faltantes

En este dataset de automóviles, identificamos varios valores faltantes que requieren atención especializada:

### Variables Numéricas con Valores Faltantes

1. **normalized-losses**: Pérdidas normalizadas (probablemente relacionadas con seguros).
   - **Estrategia**: Rellenar con la media.
   - **Justificación**: Es una medida continua donde la media representa un valor razonable.

2. **bore**: Diámetro del cilindro del motor.
   - **Estrategia**: Rellenar con la media.
   - **Justificación**: Especificación técnica continua.

3. **stroke**: Carrera del pistón.
   - **Estrategia**: Rellenar con la media.
   - **Justificación**: Parámetro técnico continuo.

4. **horsepower**: Caballos de fuerza.
   - **Estrategia**: Rellenar con la media.
   - **Justificación**: Medida de rendimiento crítica.

5. **peak-rpm**: Revoluciones por minuto pico.
   - **Estrategia**: Rellenar con la media.
   - **Justificación**: Especificación técnica.

6. **price**: Precio de venta.
   - **Estrategia**: Rellenar con la media.
   - **Justificación**: Variable objetivo potencial; la media es un estimador conservador.

### Variables Categóricas con Valores Faltantes

1. **num-of-doors**: Número de puertas.
   - **Estrategia**: Rellenar con la moda (valor más frecuente).
   - **Justificación**: Variable categórica donde la moda representa la opción más común.

## Eliminación de Columnas Irrelevantes

Además del tratamiento de valores faltantes, eliminaremos columnas que no aportan valor significativo al análisis:

1. **symboling**: Clasificación de riesgo de seguro.
   - **Razón**: Muy específica del dominio de seguros, no generalizable.

2. **engine-location**: Ubicación del motor.
   - **Razón**: En la mayoría de los casos es "front", proporcionando poca variabilidad.

## Consideraciones Éticas y Prácticas

- **Transparencia**: Documentamos todas las transformaciones realizadas.
- **Justificación**: Cada decisión está basada en el conocimiento del dominio.
- **Reversibilidad**: Mantenemos el dataset original intacto para referencia.
- **Validación**: Verificaremos que las imputaciones no distorsionen las distribuciones.

## Implementación Técnica

En la siguiente celda de código, aplicaremos:

1. **Imputación de medias**: Para variables numéricas faltantes.
2. **Imputación de moda**: Para variables categóricas faltantes.
3. **Eliminación de columnas**: Remover atributos irrelevantes.
4. **Verificación**: Mostrar el dataset resultante.

## Impacto Esperado

Después de esta limpieza, esperamos:

- Un dataset completo sin valores faltantes.
- Variables más relevantes para análisis posteriores.
- Mantenimiento de la integridad estadística de los datos.
- Preparación óptima para etapas de transformación y modelado.

Esta fase de limpieza establece una base sólida para los análisis subsiguientes, asegurando que trabajemos con datos de alta calidad y confiabilidad.

```python
# Paso 3: Limpieza de datos
# Rellenar valores faltantes en columnas numéricas con la media
df.fillna({
    "normalized-losses": df["normalized-losses"].mean(),
    "bore": df["bore"].mean(),
    "stroke": df["stroke"].mean(),
    "horsepower": df["horsepower"].mean(),
    "peak-rpm": df["peak-rpm"].mean(),
    "price": df["price"].mean()
}, inplace=True)

# Rellenar valores faltantes en columnas categóricas con el valor más frecuente
df.fillna({"num-of-doors": df["num-of-doors"].mode()[0]}, inplace=True)

# Eliminar columnas irrelevantes
columns_to_drop = ["symboling", "engine-location"]
df.drop(columns=columns_to_drop, errors='ignore', inplace=True)

print("\nDatos después de la limpieza:")
df
```

# Paso 4: Transformación de Datos - Creando Características Derivadas y Enriqueciendo el Dataset

## Conceptos de Ingeniería de Características

La transformación de datos va más allá de la limpieza; implica crear nuevas variables que capturen información adicional o representen los datos de formas más útiles para el análisis y modelado. Esta fase, conocida como "ingeniería de características" o "feature engineering", es donde convertimos datos crudos en representaciones más informativas.

### Beneficios de la Transformación de Datos

- **Mejor representación**: Variables nuevas pueden capturar relaciones complejas.
- **Reducción de dimensionalidad**: Combinar variables relacionadas.
- **Mejora del rendimiento**: Modelos pueden aprender patrones más efectivos.
- **Interpretabilidad**: Nuevas variables pueden ser más intuitivas.

## Transformaciones Específicas que Aplicaremos

### 1. Creación de Categorías de Precio

**Variable Nueva**: `Categoria_Precio`

**Justificación**:
- Los precios absolutos pueden ser difíciles de interpretar en contextos categóricos.
- Las categorías permiten segmentación de mercado más clara.
- Facilita análisis comparativos entre grupos de precios.

**Metodología**:
- Utilizaremos `pd.cut()` para crear bins discretos.
- Definiremos tres categorías: 'Bajo', 'Medio', 'Alto'.
- Límites: [0, 10000), [10000, 20000), [20000, 40000)
- Esto cubre el rango típico de precios de automóviles.

**Aplicaciones**:
- Análisis de mercado por segmentos de precio.
- Modelos de clasificación basados en rangos de precio.
- Visualizaciones más intuitivas.

### 2. Cálculo del Consumo Combinado

**Variable Nueva**: `consumo_combinado`

**Justificación**:
- El consumo de combustible varía entre ciudad y carretera.
- Un promedio proporciona una medida general del rendimiento.
- Simplifica comparaciones entre vehículos.

**Metodología**:
- Fórmula: `(city-mpg + highway-mpg) / 2`
- Promedio aritmético simple de ambos consumos.
- Mantiene la unidad de medida (millas por galón).

**Aplicaciones**:
- Comparaciones directas de eficiencia energética.
- Análisis de impacto ambiental.
- Optimización de flotas vehiculares.

## Consideraciones Técnicas

### Validación de Existencia de Columnas
- Verificaremos que las columnas base ('price', 'city-mpg', 'highway-mpg') existan antes de crear las nuevas.
- Esto previene errores si el dataset ha sido modificado previamente.

### Tipos de Datos Adecuados
- `Categoria_Precio`: Variable categórica ordinal.
- `consumo_combinado`: Variable numérica continua.

### Impacto en el Análisis
- Aumenta el número de variables disponibles para modelado.
- Proporciona perspectivas adicionales sobre los datos.
- Puede revelar patrones no evidentes en las variables originales.

## Implementación en Código

En la siguiente celda, ejecutaremos:

1. **Creación de categorías de precio**: Usando pandas cut con bins definidos.
2. **Cálculo del consumo combinado**: Operación aritmética simple.
3. **Verificación de resultados**: Visualización del dataset transformado.

## Validación de las Transformaciones

Después de aplicar estas transformaciones, verificaremos:

- Que las nuevas columnas se crearon correctamente.
- Que los valores están en rangos esperados.
- Que no se introdujeron valores faltantes nuevos.
- Que las distribuciones de las nuevas variables son lógicas.

## Próximos Pasos

Estas nuevas características enriquecerán nuestro análisis en las fases de visualización y modelado. La categoría de precio permitirá segmentaciones más naturales, mientras que el consumo combinado proporcionará una medida unificada de eficiencia.

Esta transformación marca la transición de datos limpios a datos enriquecidos, preparados para insights más profundos y modelado predictivo avanzado.

```python
# Paso 4: Transformación de datos
# Crear una nueva columna categórica basada en el precio
if 'price' in df.columns:
    df['Categoria_Precio'] = pd.cut(df['price'], bins=[0, 10000, 20000, 40000], labels=['Bajo', 'Medio', 'Alto'], right=False)

# Crear una nueva columna con el consumo combinado (promedio de ciudad y carretera)
df['consumo_combinado'] = (df['city-mpg'] + df['highway-mpg']) / 2

print("\nDatos después de la transformación:")
df
```
