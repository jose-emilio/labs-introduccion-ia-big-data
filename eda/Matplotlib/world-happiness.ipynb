# Análisis Exploratorio de Datos: World Happiness Report

Este notebook realiza un análisis exploratorio de datos (EDA) utilizando el dataset del World Happiness Report de 2015 a 2021. El objetivo es visualizar y comprender las tendencias en la felicidad y sus factores asociados en varios países, con un enfoque en España, Francia y Portugal.

## Contexto del Dataset
El World Happiness Report es una publicación anual que clasifica a los países según su nivel de felicidad basado en encuestas y datos económicos. Incluye métricas como:
- **Happiness Score**: Puntuación general de felicidad.
- **Economy (GDP per Capita)**: Producto Interior Bruto per cápita.
- **Family (Social Support)**: Apoyo social.
- **Health (Life Expectancy)**: Esperanza de vida.
- **Freedom**: Libertad percibida.
- **Trust (Government Corruption)**: Confianza en el gobierno (inverso a corrupción).
- **Generosity**: Generosidad.

Utilizaremos Matplotlib para crear visualizaciones que nos ayuden a identificar patrones, tendencias y correlaciones.

## Ejercicio 1: Evolución de la Puntuación de Felicidad

En este ejercicio, creamos un gráfico de líneas para visualizar la evolución temporal de la puntuación de felicidad en España, Francia, Portugal e Italia desde 2015 hasta 2021.

### Código Analizado
- **Importaciones**: Utilizamos `pandas` para manipular los datos y `matplotlib.pyplot` para la visualización.
- **Filtrado de Datos**: Seleccionamos los registros correspondientes a cada país usando `report['Country'] == 'País'`.
- **Gráfico de Líneas**: 
  - `plt.plot()` dibuja líneas para cada país con estilos diferentes: España (rojo con círculos y líneas discontinuas), Francia (azul con estrellas y puntos), Portugal (verde con diamantes y líneas punteadas), Italia (magenta con triángulos y líneas punteadas).
  - Etiquetas en ejes (`xlabel`, `ylabel`), título y leyenda para claridad.
  - Anotación para el impacto de COVID-19 en 2020.
- **Visualización**: `plt.show()` muestra el gráfico.

### Interpretación de Resultados
Este gráfico permite comparar las tendencias de felicidad entre los cuatro países ibéricos. Observamos fluctuaciones anuales que pueden estar influenciadas por eventos económicos, políticos o sociales. Por ejemplo, caídas en ciertos años podrían correlacionarse con crisis económicas o pandemias.

```python
import pandas as pd
import matplotlib.pyplot as plt

report = pd.read_csv('world-happiness-report.csv')

spain = report[(report['Country'] == 'Spain')]
france = report[(report['Country'] == 'France')]
portugal = report[(report['Country'] == 'Portugal')]
italy = report[(report['Country'] == 'Italy')]

plt.plot(spain.Year, spain['Happiness Score'], 'ro--', label='Spain')
plt.plot(france.Year, france['Happiness Score'], 'b*:', label='France')
plt.plot(portugal.Year, portugal['Happiness Score'], 'gd-.', label='Portugal')
plt.plot(italy.Year, italy['Happiness Score'], 'm^-', label='Italy')

plt.xlabel('Year')
plt.ylabel('Happiness Score')
plt.title('Happiness Score Evolution')
plt.legend()

# Anotación para evento global (ej. COVID-19 impacto en 2020)
plt.annotate('COVID-19 Impact', xy=(2020, spain[spain.Year == 2020]['Happiness Score'].values[0]), xytext=(2018, 6.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

## Ejercicio 2: Evolución del PIB per Cápita

Este ejercicio utiliza un gráfico de barras para mostrar la evolución del Producto Interior Bruto (PIB) per cápita en España, Francia y Portugal.

### Código Analizado
- **Filtrado**: Similar al ejercicio anterior, filtramos por país.
- **Gráfico de Barras Agrupadas**: 
  - Usamos `plt.bar()` con desplazamientos en el eje x (`Year - 0.2`, `Year`, `Year + 0.2`) para evitar superposición.
  - Cada país tiene un color único (rojo, azul, verde).
  - Ancho de barras de 0.2 para ajuste.
- **Líneas de Tendencia**: Ajuste polinomial de grado 1 para mostrar la tendencia general por país.
- **Elementos del Gráfico**: Etiquetas, título y leyenda centrada.

### Interpretación de Resultados
Las barras permiten comparar el PIB per cápita año a año y entre países. Un crecimiento consistente indica estabilidad económica, mientras que variaciones pueden reflejar recesiones o expansiones.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

report = pd.read_csv('world-happiness-report-2015-2021-cleaned.csv')

spain = report[(report['Country'] == 'Spain')]
france = report[(report['Country'] == 'France')]
portugal = report[(report['Country'] == 'Portugal')]

plt.bar(spain.Year - 0.2, spain['Economy (GDP per Capita)'], width=0.2, color='r', label='Spain')
plt.bar(france.Year, france['Economy (GDP per Capita)'], width=0.2, color='b', label='France')
plt.bar(portugal.Year + 0.2, portugal['Economy (GDP per Capita)'], width=0.2, color='g', label='Portugal')

# Líneas de tendencia
for country, data, color in [('Spain', spain, 'r'), ('France', france, 'b'), ('Portugal', portugal, 'g')]:
    z = np.polyfit(data.Year, data['Economy (GDP per Capita)'], 1)
    p = np.poly1d(z)
    plt.plot(data.Year, p(data.Year), color + '--', linewidth=2, label=f'{country} Trend')

plt.xlabel('Year')
plt.ylabel('GDP per Capita')
plt.title('GDP per Capita Evolution')
plt.legend(loc='center')

plt.show()
```

## Ejercicio 4: Distribución por Categorías de Felicidad

Creamos gráficos de tarta para categorizar la felicidad en 'LOW', 'MED', 'HIGH' para 2018 y 2019.

### Código Analizado
- **Categorización**: Usamos `pd.qcut()` para dividir las puntuaciones en terciles basados en la distribución estadística: LOW (tercil inferior), MED (medio), HIGH (superior).
- **Agrupación**: `groupby()` cuenta países por categoría.
- **Gráficos de Tarta**: 
  - `plt.pie()` con etiquetas, porcentajes automáticos (`autopct`), y colores consistentes (rojo, verde, azul).
  - Dos subplots lado a lado.

### Interpretación de Resultados
Los gráficos muestran la proporción de países en cada categoría basada en terciles. Esto asegura una distribución equilibrada por definición, reflejando la variabilidad real en los datos.

### Comparación con Histogramas
A diferencia de los histogramas continuos, las tartas ofrecen una vista categórica, facilitando la comprensión de grupos.

```python
import pandas as pd
import matplotlib.pyplot as plt

report = pd.read_csv('world-happiness-report-2015-2021-cleaned.csv')

labels=['LOW', 'MED', 'HIGH']

# Usar cuantiles para categorías basadas en distribución estadística
report['Range']=pd.qcut(report['Happiness Score'], 3, labels=labels)

report2018 = report[(report['Year'] == 2018)].groupby(by=['Range']).Country.agg(['count'])
report2019 = report[(report['Year'] == 2019)].groupby(by=['Range']).Country.agg(['count'])

def value(x):
    return str(round(x)) + "%"

plt.subplot(1,2,1)
plt.pie(report2018['count'], labels=labels, autopct=value, colors=['r','g','b'])
plt.title('2018')

plt.subplot(1,2,2)
plt.pie(report2019['count'], labels=labels, autopct=value, colors=['r','g','b'])
plt.title('2019')

plt.show()
```

## Ejercicio 5: Combinación de Tartas e Histogramas

Este ejercicio combina los gráficos de tarta y histogramas en un layout de 2x2, con colores consistentes.

### Código Analizado
- **Subplots 2x2**: Arriba tartas (2018 y 2019), abajo histogramas coloreados por categoría.
- **Colores Consistentes**: Rojo para LOW, verde para MED, azul para HIGH.
- **Histogramas Apilados**: Múltiples `plt.hist()` para cada categoría en el mismo subplot, con transparencias (`alpha=0.7`) y leyendas.

### Interpretación de Resultados
La combinación permite ver tanto la proporción categórica (tartas) como la distribución continua dentro de cada categoría (histogramas). Los colores vinculan visualmente las partes.

### Beneficios
Facilita la comparación entre representaciones categóricas y continuas, proporcionando una vista más completa de los datos.

```python
import pandas as pd
import matplotlib.pyplot as plt

report = pd.read_csv('world-happiness-report-2015-2021-cleaned.csv')

labels=['LOW', 'MED', 'HIGH']

report['Range']=pd.cut(report['Happiness Score'], [0,4,6,8], labels=labels)

report2018 = report[(report['Year'] == 2018)].groupby(by=['Range']).Country.agg(['count'])
report2019 = report[(report['Year'] == 2019)].groupby(by=['Range']).Country.agg(['count'])

year2018 = report[(report['Year'] == 2018)]
year2019 = report[(report['Year'] == 2019)]

year2018_low = year2018[(year2018['Range'] == 'LOW')]
year2018_med = year2018[(year2018['Range'] == 'MED')]
year2018_high = year2018[(year2018['Range'] == 'HIGH')]

year2019_low = year2019[(year2019['Range'] == 'LOW')]
year2019_med = year2019[(year2019['Range'] == 'MED')]
year2019_high = year2019[(year2019['Range'] == 'HIGH')]

def value(x):
    return str(round(x)) + "%"

plt.subplot(2,2,1)
plt.pie(report2018['count'], labels=labels, autopct=value, colors=['r','g','b'])
plt.title('2018')

plt.subplot(2,2,2)
plt.pie(report2019['count'], labels=labels, autopct=value, colors=['r','g','b'])
plt.title('2019')

plt.subplot(2,2,3)
plt.hist(year2018_low['Happiness Score'], edgecolor='black', color='r', alpha=0.7, label='LOW' )
plt.hist(year2018_med['Happiness Score'], edgecolor='black', color='g', alpha=0.7, label='MED' )
plt.hist(year2018_high['Happiness Score'], edgecolor='black', color='b', alpha=0.7, label='HIGH' )
plt.xlabel('Happiness Score')
plt.ylabel('Countries')
plt.legend()

plt.subplot(2,2,4)
plt.hist(year2019_low['Happiness Score'], edgecolor='black', color='r', alpha=0.7, label='LOW' )
plt.hist(year2019_med['Happiness Score'], edgecolor='black', color='g', alpha=0.7, label='MED' )
plt.hist(year2019_high['Happiness Score'], edgecolor='black', color='b', alpha=0.7, label='HIGH' )
plt.xlabel('Happiness Score')
plt.legend()

plt.show()
```

## Ejercicio 6: Gráficos de Dispersión con Variables Relacionadas

Creamos 6 gráficos de dispersión para explorar correlaciones entre la felicidad y otros factores, con líneas de regresión.

### Código Analizado
- **Subplots 3x2**: Seis gráficos en una cuadrícula generados con un bucle.
- **Variables**: GDP per Capita, Social Support, Life Expectancy, Freedom, Government Corruption, Generosity.
- **Dispersión**: `plt.scatter()` con puntos pequeños (`s=0.2`) para manejar muchos datos.
- **Líneas de Regresión**: Ajuste lineal usando `np.polyfit()` para cuantificar tendencias.
- **Layout**: `plt.tight_layout()` para evitar superposiciones.

### Interpretación de Resultados
Los gráficos de dispersión muestran relaciones potenciales. Por ejemplo:
- Correlación positiva entre GDP y felicidad (países ricos tienden a ser más felices).
- Social Support y Life Expectancy también muestran tendencias positivas.
- Freedom y Generosity pueden tener correlaciones más débiles.

### Análisis de Correlación
Identificar patrones lineales o no lineales ayuda a entender qué factores influyen más en la felicidad. Las líneas rojas punteadas representan el ajuste lineal.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

report = pd.read_csv('world-happiness-report-2015-2021-cleaned.csv')

variables = [
    ('Economy (GDP per Capita)', 'GDP per Capita'),
    ('Family (Social Support)', 'Social Support'),
    ('Health (Life Expectancy)', 'Life Expectancy'),
    ('Freedom', 'Freedom'),
    ('Trust (Government Corruption)', 'Government Corruption'),
    ('Generosity', 'Generosity')
]

for i, (col, label) in enumerate(variables, 1):
    plt.subplot(3,2,i)
    plt.scatter(report[col], report['Happiness Score'], s=0.2, label='Data Points')
    plt.xlabel(label)
    if i % 2 == 1:
        plt.ylabel('Happiness Score')
    
    # Línea de regresión
    z = np.polyfit(report[col], report['Happiness Score'], 1)
    p = np.poly1d(z)
    x_range = np.linspace(report[col].min(), report[col].max(), 100)
    plt.plot(x_range, p(x_range), 'r--', linewidth=1, label='Regression Line')
    plt.legend()

plt.tight_layout()
plt.show()
```
